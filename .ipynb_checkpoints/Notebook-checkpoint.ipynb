```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv("./supermarket_sales - Sheet1.xls")
df.head()
```

```python
df.describe()
```

```python
df.drop(["Invoice ID"],axis = 1, inplace = True)
```

```python
df.head()
```

### 1. Percentage of member and non member 
<p>
    This will tell us the distribution of total customers as members or non-members along with their gender.
    We can then further analyse which of these 4 catagories brings in the most revenue and plan on retaining such members.
</p>


```python
fig = px.bar(df,x = "Customer type", color = "Gender",barmode='group',title = "Percentage Members based on gender of the customers")
fig.show()
```

### 2.Product based on product line
<p>
    This will help us find what products among- Fashion accessories, food and beverages, Electronic accesories, sports and travel, home and lifestyle or health and beauty sell the most from all the branches in terms of sheer quantity.

</p>

```python
df['Product line'].value_counts()
```

```python
# check
fig = px.pie(df, values='Quantity', names='Product line',title="Share of different product lines sold at the supermarket from all its branches")
fig.show()
```

### 3. branch wise sale 
<p>
    We can find out the most profitable branch for the supermarket.
</p>

```python
branchWise_df = df.groupby('Branch')['Total'].sum()
branchWise_df = branchWise_df.to_frame().reset_index()
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(branchWise_df['Branch'], branchWise_df['Total'],width = 0.2)
 
plt.xlabel("Branch")
plt.ylabel("Total revenue generated")
plt.title("Total revenue generated by different branches")
```

### 4. Total revenue generated by different product line
<p>
We will know what category of products sells more from al that offered in all stores of the supermarket chain.    
</p>

```python
productWise = df.groupby('Product line')['Total'].sum()
productWise = productWise.to_frame().reset_index()
type(branchWise_df) 
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(productWise['Product line'], productWise['Total'],width = 0.2)
 
plt.xlabel("Products")
plt.ylabel("Total revenue generated")
plt.title("Total revenue generated by different product line")


```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(productWise['Total'], labels = productWise['Product line'],textprops={'color':"w"})
plt.show()
```

### 5. Spread of sales based on date and time.

```python
df.head()
```

```python
df['Date'][1][2]
```

```python
def quarters(x):
    if x in range(1,4):
        return "1st Quarter"
    if x in range(4,7):
        return "2nd Quarter"
    if x in range(7,10):
        return "3rd Quarter"
    if x in range(10,13):
        return "4rt Quarter"
    
# df['x_range']=table_to_plot['x_variable']
# table_to_plot['y_range'] = table_to_plot['y_variable']
quarter_df = pd.DataFrame(columns=['date','quarters'])
quarter_df['date'] = df['Date'][1][2]
quarter_df['quarters'] = df['Date'].apply(lambda x: quarters(df['Date'][:][2]))

quarter_df
```
